```python
import speedtest
import time

def test_speed():
    st = speedtest.Speedtest()
    download_speed = st.download()
    upload_speed = st.upload()
    ping_speed = st.results.ping
    cont = 1
      
    while (cont <=100):
        print("Teste ", cont)
        cont +=1
        print(f"Velocidade de download: {download_speed / 10**6:.2f} Mbps")
        print(f"Velocidade de upload: {upload_speed / 10**6:.2f} Mbps")
        print(f"Latência de ping: {ping_speed:.2f} ms")
        print(" ")
    

while True:
    test_speed()
    time.sleep(1)
```
